```python
import pandas as pd
```

```python
df = pd.read_csv("joy3.csv")
df
```

```python
df1 = df[['date', 'timestamp', 'price','polarity.score', 'negative.score', 'neutral.score', 'positive.score', 'Price_Change']]
df1
```

```python
df2 = pd.read_csv("joy4.csv")
df2
```

```python
df3 = df2[['date', 'timestamp', 'price', 'polarity.score','negative.score', 'neutral.score', 'positive.score', 'Price_Change']]
df3
```

```python
total_df = [df1, df3]
```

```python
total_df2 = pd.concat(total_df, ignore_index=True)
total_df2_sort = total_df2.sort_values(by='date')
```

<h3>Price and Scores by Hour</h3>

```python
total_df3= total_df2_sort[['timestamp', 'price']]
```

```python
total_df4= total_df2_sort[['timestamp', 'polarity.score']]
```

```python
total_df5= total_df2_sort[['timestamp', 'positive.score']]
```
